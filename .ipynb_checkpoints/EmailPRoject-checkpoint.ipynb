```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string
import re

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

def transform_text(text):
    text = text.lower()
    text = nltk.word_tokenize(text)
    
    # Retaining only numbers and alphabets
    y = []
    for i in text:
        if i.isalnum():
            y.append(i)
    
    # Clone copy of list y
    text = y[:]
    y.clear()
    
    # Stopwords removal and removal of special characters/punctuation
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
            
    text = y[:]
    y.clear()
    
    # Lemmatization
    for i in text:
        y.append(lemmatizer.lemmatize(i)) 
        
    return " ".join(text)
```

### Balance The Dataset
Balance the dataset : trong machine learning có nghĩa là điều chỉnh tập dữ liệu để có số lượng quan sát tương đương giữa các lớp (classes). Trong một tập dữ liệu không cân bằng, một lớp có số lượng quan sát nhiều hơn lớp khác, gây ra sự mất cân đối và có thể ảnh hưởng đến hiệu suất của mô hình.
+ Upsampling
+ Downsampling

```python
from sklearn.utils import resample

# create two different dataframe of majority and minority class 

df_majority = df[(df['label'] == 0)] 

df_minority = df[(df['label'] == 1)] 

# upsample minority class
df_minority_upsampled = resample(df_minority,
                                 
replace = True,    # sample with replacement  
                                 
 n_samples = 4515, # to match majority class     
                                 
 random_state = 42) 

# reproducible results
    
# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
```

### Encode Independent Variable(V2)

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

# Create a CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df["message"])
y = df["label"]

# Convert the count matrix to an array
count_array = X.toarray()

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(count_array, y, test_size=0.3, random_state=0)

# Create a DataFrame from the count array
df1 = pd.DataFrame(data=count_array, columns=vectorizer.get_feature_names_out())
df1
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.metrics import precision_recall_fscore_support
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

# Create a Random Forest classifier
rf_model = RandomForestClassifier(n_estimators=100, criterion='gini')

# Train the model
rf_model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = rf_model.predict(X_test)

# Calculate the accuracy score
accuracy_rf = accuracy_score(y_test, y_pred)
print("Test Accuracy:", accuracy_rf)



# Generate the classification report
report = classification_report(y_test, y_pred)
print("Confusion Matrix Report:")
print(report)


# Perform cross-validation
scores = cross_val_score(rf_model, X, y, cv=10)

# Print the accuracy scores for each fold
print("Cross-Validation Accuracy Scores:", scores)

# Calculate the mean accuracy score across all folds
mean_accuracy = scores.mean()
print("Mean Accuracy:", mean_accuracy)



# Calculate precision, recall, and f1-score
precision_rf, recall_rf, f1_rf, _ = precision_recall_fscore_support(y_test, y_pred, average='weighted')

print("Precision Score:", precision_rf)
print("Recall Score:", recall_rf)
print("F1 Score:", f1_rf)
```

```python
import numpy as np

accuracy_scores = [accuracy_rf, accuracy_lg, accuracy_svc, accuracy_nb]
algorithms = ['Random Forest', 'Logistic Regression', 'SVM', 'Naive Bayes']

# Convert accuracy scores to percentages
accuracy_scores = [score * 100 for score in accuracy_scores]

# Define colors for each algorithm
colors = ['green', 'blue', 'orange', 'purple']

# Create the bar plot
fig, ax = plt.subplots()

# Adjust the width of the bars for better visualization
bar_width = 0.6

bars = ax.bar(algorithms, accuracy_scores, width=bar_width, edgecolor='black')

# Assign colors to each bar
for bar, color in zip(bars, colors):
    bar.set_color(color)

# Add percentage labels on each bar
for bar, score in zip(bars, accuracy_scores):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
            f'{score:.1f}%', ha='center', va='bottom')

plt.xlabel('Algorithms')
plt.ylabel('Accuracy Score (%)')
plt.title('Comparison of Algorithm Accuracy')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Convert accuracy scores to percentages
precision_scores = [precision_rf * 100, precision_lg * 100, precision_svc * 100, precision_nb * 100]
recall_scores = [recall_rf * 100, recall_lg * 100, recall_svc * 100, recall_nb * 100]
f1_scores = [f1_rf * 100, f1_lg * 100, f1_svc * 100, f1_nb * 100]
algorithms = ['Random Forest', 'Logistic Regression', 'SVM', 'Naive Bayes']

# Create an array of indices for x-axis
x = np.arange(len(algorithms))

# Set the width of the bars
bar_width = 0.2

# Create the bar plots for Precision, Recall, and F1-score
plt.bar(x - bar_width, precision_scores, width=bar_width, label='Precision', color='steelblue')
plt.bar(x, recall_scores, width=bar_width, label='Recall', color='mediumseagreen')
plt.bar(x + bar_width, f1_scores, width=bar_width, label='F1-score', color='indianred')

# Set the x-axis ticks and labels
plt.xticks(x, algorithms)

# Set the y-axis label
plt.ylabel('Scores (%)')

# Set the title
plt.title('Precision, Recall, and F1-score Comparison')

# Add percentage labels above each bar
for i, score in enumerate(precision_scores):
    plt.text(i - bar_width, score + 2, f'{score:.2f}%', color='black', ha='center')
for i, score in enumerate(recall_scores):
    plt.text(i, score + 2, f'{score:.2f}%', color='black', ha='center')
for i, score in enumerate(f1_scores):
    plt.text(i + bar_width, score + 2, f'{score:.2f}%', color='black', ha='center')

# Add a legend
plt.legend()

# Add gridlines
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Set the y-axis limits
plt.ylim([0, max(max(precision_scores), max(recall_scores), max(f1_scores)) + 5])

# Set the figure size
plt.figure(figsize=(10, 6))

# Show the plot
plt.show()
```
